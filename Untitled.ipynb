```python

# K-MEANS CLUSTERING COMPLETO


#  Importar libreras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

#  Cargar dataset
ruta = '/Users/juandavidrodriguez/Desktop/Mall_Customers 2.csv'
df = pd.read_csv(ruta)

#  Preprocesamiento
df_clean = df.drop(columns=['CustomerID'])
df_clean['Gender'] = df_clean['Gender'].map({'Male': 0, 'Female': 1})

scaler = StandardScaler()
X = scaler.fit_transform(df_clean)

#  Mtodo del codo para seleccionar K
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.title('Mtodo del codo - KMeans')
plt.xlabel('Nmero de clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()

# Entrenar modelo con K=5
kmeans = KMeans(n_clusters=5, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

#  Evaluacin del modelo
print("Silhouette Score:", silhouette_score(X, labels_kmeans))
print("Calinski-Harabasz Index:", calinski_harabasz_score(X, labels_kmeans))
print("Davies-Bouldin Index:", davies_bouldin_score(X, labels_kmeans))

#  Visualizacin
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_kmeans, cmap='rainbow')
plt.title('Clusters K-Means visualizados con PCA')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.grid(True)
plt.show()
```

```python

# CLUSTERING JERRQUICO COMPLETO

#  Importar libreras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA

#  Cargar dataset
ruta = '/Users/juandavidrodriguez/Desktop/Mall_Customers 2.csv'
df = pd.read_csv(ruta)

# Preprocesamiento
df_clean = df.drop(columns=['CustomerID'])
df_clean['Gender'] = df_clean['Gender'].map({'Male': 0, 'Female': 1})

scaler = StandardScaler()
X = scaler.fit_transform(df_clean)

#  Dendrograma para observar agrupaciones
linked = linkage(X, method='ward')
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title('Dendrograma - Clustering Jerrquico')
plt.xlabel('Clientes')
plt.ylabel('Distancia')
plt.grid(True)
plt.show()

# Entrenar modelo con n_clusters=5
hierarchical = AgglomerativeClustering(n_clusters=5)
labels_hierarchical = hierarchical.fit_predict(X)

# Evaluacin del modelo
print("Silhouette Score:", silhouette_score(X, labels_hierarchical))
print("Calinski-Harabasz Index:", calinski_harabasz_score(X, labels_hierarchical))
print("Davies-Bouldin Index:", davies_bouldin_score(X, labels_hierarchical))

#  Visualizacin con PCA
X_pca = PCA(n_components=2).fit_transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_hierarchical, cmap='rainbow')
plt.title('Clusters Jerrquicos visualizados con PCA')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.grid(True)
plt.show()
```
